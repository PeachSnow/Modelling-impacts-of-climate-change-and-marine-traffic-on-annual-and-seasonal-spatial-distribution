```python
def myplot(score,coeff,labels=None):
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]
    scalex = 1.0/(xs.max() - xs.min())
    scaley = 1.0/(ys.max() - ys.min())
    plt.scatter(xs * scalex,ys * scaley)
    for i in range(n):
        plt.arrow(0, 0, coeff[i,0], coeff[i,1],color = 'r',alpha = 0.5)
        if labels is None:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, "Var"+str(i+1), color = 'g', ha = 'center', va = 'center')
        else:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, labels[i], color = 'g', ha = 'center', va = 'center')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.xlabel("PC{}".format(1))
plt.ylabel("PC{}".format(2))
plt.grid()

#Call the function. Use only the 2 PCs.
myplot(X_train[:,0:2],np.transpose(pca.components_[0:2, :]))
plt.show()
```

# PCA - Ionio - RUN2

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid", palette="ch:r=-.2,d=.3_r")
%matplotlib inline
import warnings
warnings.filterwarnings('ignore') #https://www.kaggle.com/code/gowtham0312/house-rent-estimation-with-randomforest

data = 'IonioRUN2.csv'
df = pd.read_csv('IonioRUN2.csv', header=None)
# preview the dataset


col_names = ['NEAR_DIST', 'BATmedian', 'HGmedian','dist40m','Grid_x','Grid_y','Dist-100','Dist-200','Dist-500','Dist-1000','Dist-1500','DistCanyou','LENGTH_KM','SST', 'CHL','Scoeruleoalba_Presence']


df = df.iloc[1: , :]
df.columns = col_names
col_names

df.head()
```

```python
df = df.drop(['LENGTH_KM'], axis=1)
df.isnull().sum()
```

```python
df = df.astype(float)
print(df.dtypes)
```

```python
X = df.drop(['Scoeruleoalba_Presence'], axis=1)
y = df['Scoeruleoalba_Presence']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

from sklearn.decomposition import PCA

pcaIonioRun2 = PCA()
X_train = pcaIonioRun2.fit_transform(X_train)
X_test = pcaIonioRun2.transform(X_test)

explained_variance = pcaIonioRun2.explained_variance_ratio_
cum_sum_eigenvalues = np.cumsum(explained_variance)

plt.bar(range(0,len(explained_variance)), explained_variance, alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(0,len(cum_sum_eigenvalues)), cum_sum_eigenvalues, where='mid',label='Cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
print(pd.DataFrame(pcaIonioRun2.components_,columns=X.columns[0:],index = ['PC-1','PC-2','PC-3','PC-4','PC-5','PC-6','PC-7','PC-8','PC-9','PC-10','PC-11','PC-12','PC-13', 'PC-14']))
```

```python
print(abs(pcaIonioRun2.components_ ))
```

```python
def myplot(score,coeff,labels=None):
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]
    scalex = 1.0/(xs.max() - xs.min())
    scaley = 1.0/(ys.max() - ys.min())
    plt.scatter(xs * scalex,ys * scaley)
    for i in range(n):
        plt.arrow(0, 0, coeff[i,0], coeff[i,1],color = 'r',alpha = 0.5)
        if labels is None:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, "Var"+str(i+1), color = 'g', ha = 'center', va = 'center')
        else:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, labels[i], color = 'g', ha = 'center', va = 'center')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.xlabel("PC{}".format(1))
plt.ylabel("PC{}".format(2))
plt.grid()

#Call the function. Use only the 2 PCs.
myplot(X_train[:,0:2],np.transpose(pcaIonioRun2.components_[0:2, :]))
plt.show()
```

```python
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
np.random.seed(0)

#X_train = pca.fit_transform(X_train)
#X_test = pca.transform(X_test)

# number of components
n_pcs= pcaIonioRun2.components_.shape[0]

# get the index of the most important feature on EACH component
# LIST COMPREHENSION HERE
most_important = [np.abs(pcaIonioRun2.components_[i]).argmax() for i in range(n_pcs)]

initial_feature_names = df.columns
# get the names
most_important_names = [initial_feature_names[most_important[i]] for i in range(n_pcs)]

# LIST COMPREHENSION HERE AGAIN
dic = {'PC{}'.format(i): most_important_names[i] for i in range(n_pcs)}

# build the dataframe
dfimportance = pd.DataFrame(dic.items())

dfimportance
```
